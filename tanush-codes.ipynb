```python
train_df = pd.read_csv("/kaggle/input/practice-hackathon/Train.csv")
```

```python
## Label encoding column A of our dataset
from sklearn import preprocessing
  
# label_encoder object knows 
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
train_df['A']= label_encoder.fit_transform(train_df['A'])
  
train_df['A'].unique()
```

```python
X_all = train_df.drop(['Target'], axis=1)
y_all = train_df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.metrics import MeanSquaredError
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import StandardScaler

# Normalize the input data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

classifier = Sequential()

# Adding the input layer and the first hidden layer

classifier.add(InputLayer(input_shape = (8,)))

classifier.add(Dense(units=64, activation='relu', kernel_regularizer=l2(0.075)))

# Adding the second hidden layer
classifier.add(Dense(units=32, activation='relu', kernel_regularizer=l2(0.075)))

# Adding the third hidden layer
classifier.add(Dense(units=16, activation='relu', kernel_regularizer=l2(0.075)))

# Adding the output layer
classifier.add(Dense(units=1))

# Compiling the ANN
classifier.compile(loss='mse', optimizer='adamax')

xtrain, xval, ytrain, yval = train_test_split(X_train, y_train, test_size = 0.33, random_state = 42)



# Fit the model with normalized data
model_history = classifier.fit(xtrain, ytrain, validation_split=0.2, batch_size=64, epochs=1250, verbose=1, validation_data=(xval, yval))

```

```python
ann_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(ann_pred, y_test))
```
